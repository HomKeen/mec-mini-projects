```python
from plotnine import *
```

```python
from plotnine.data import mtcars
```

```python
mtcars.head()
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg')) 
        + geom_point()
        + theme_bw())
```

```python
(ggplot(mtcars, 
            aes('wt', 'mpg', color='factor(gear)'))
            + geom_point()
            + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', size='cyl'))
        + geom_point() 
        + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)'))
        + geom_point() 
        + facet_wrap('~cyl') 
        + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)'))
        + geom_point() 
        + stat_smooth(method='lm') 
        + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', 
            size='cyl'))
        + geom_point() 
        + facet_wrap('~am') 
        + theme_bw())
```

```python
(ggplot(mtcars, 
        aes('wt', 'mpg', 
            color='factor(gear)', 
            size='cyl'))
        + geom_point() 
        + facet_grid('am ~ carb') 
        + theme_bw())
```
